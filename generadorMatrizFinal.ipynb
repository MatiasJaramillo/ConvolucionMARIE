```python
import random

def generarMatriz(rows, cols, ran):
  sizeMatriz = rows * cols
  matrizX1D = []
  if ran == "s" :
    matrizX1D = [random.randint(1, 10) for i in range(sizeMatriz)]
  else :
    for i in range(rows) :
      for j in range(cols) :
        matrizX1D.append(j+1)
  matrizY1D = []
  for k in range(sizeMatriz):
    matrizY1D.append(0)
  print("XIndex, HEX 018\nXRows, DEC " + formatNumber(rows) + "\nXColumns, DEC " + formatNumber(cols) + "\nXMatriz, DEC " + formatNumber(sizeMatriz))
  for i in range(sizeMatriz):
    print("\tDEC " + formatNumber(matrizX1D[i]))
  print("YIndex, HEX " + getYIndex("18", sizeMatriz) + "\nYRows, DEC " + formatNumber(rows) + "\nYColumns, DEC " + formatNumber(cols) + "\nYMatriz, DEC " + formatNumber(sizeMatriz))
  for i in range(sizeMatriz):
    print("\tDEC 000")
  print("\n--------Copiar y Pegar en Java--------\n")
  generarMatrizJava(rows, cols, matrizX1D, matrizY1D)

def generarMatrizJava(rows, cols, array, array2):
    print("int[][] matrizX = {")
    for i in range(rows):
        start = i * cols
        end = start + cols
        row = array[start:end]
        row_str = ", ".join(map(str, row))
        if i == rows - 1:
            print(f"    {{{row_str}}}")
        else:
            print(f"    {{{row_str}}},")
    print("};")

    print("\nint[][] matrizY = {")
    for i in range(rows):
        start = i * cols
        end = start + cols
        row = array2[start:end]
        row_str = ", ".join(map(str, row))
        if i == rows - 1:
            print(f"    {{{row_str}}}")
        else:
            print(f"    {{{row_str}}},")
    print("};")

def formatNumber(num) :
  numStr = ""
  if num < 0:
    num *= -1
    numStr += "-"
  if num > 99:
    numStr += str(num)
  elif num < 100 and num > 9:
    numStr += "0" + str(num)
  else:
    numStr += "00" + str(num)
  return numStr

def getYIndex(xindex, size):
  decNum = int(xindex, 16)
  decNum += 4 + size
  hexNum = hex(decNum)[2:].upper()
  auxStr = ""
  while len(hexNum) < 3:
    auxStr += "0"
    hexNum = auxStr + hexNum
  return hexNum

### aqui empieza el programa *********************************

rows = int(input("Numero de filas: "))
cols = int(input("Numero de columnas: "))
ranBool = input("Deseas generar numeros aleatorios s/n: ")

print("\n--------Copiar y Pegar en MARIE--------\n")
generarMatriz(rows, cols, ranBool)
```
